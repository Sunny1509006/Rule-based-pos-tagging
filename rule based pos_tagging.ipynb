```python
def pos_tagger(sentence):
  # split the sentence into words
  words = sentence.split()

  # create a list of tuples, where each tuple contains a word and its POS tag
  tags = []
  for word in words:
    # determine the POS tag for the word
    if word.isnumeric():
      tag = "NUM"
    #coordinating conjunctions
    elif word.lower() in ['and', 'but', 'or', 'still', 'otherwise', 'so', 'yet']:
      tag = "CC"
    #Determiner
    elif word.lower() in ['the', 'a', 'an', 'that', 'these', 'this', 'those']:
      tag = "DT"
    #preposition
    elif word.lower() in ['to', 'in', 'at', 'of', 'on', 'into', 'onto', 'since', 'for', 'by', 'during', 
                         'from', 'with', 'within', 'inside', 'over', 'above', 'below', 'beneath', 'under',
                         'underneath', 'near', 'between', 'among', 'opposite', 'off', 'through', 'toward']:
      tag = "PREP"
    #Auxiliary verb
    elif word.lower() in ['am', 'is', 'are', 'was', 'were', 'being', 'been', 'be', 'has', 'have', 'had', 'having', 'does',
                         'do', 'did']:
      tag = "AX"
    #Modal auxiliary verb
    elif word.lower() in ['must', 'ought', 'should', 'might', 'would', 'shall', 'can', 'could']:
      tag = "MD"
    #WH pronouns
    elif word.lower() in ['what', 'which', 'whom', 'whose', 'who', 'when', 'where', 'Whatever', 'Whatsoever', 'Whichever',
                         'Whoever', 'Whosoever', 'Whomever', 'Whomsoever', 'Whosever', 'while', 'wherever']:
      tag = "WP"
    #personal pronoun
    elif word.lower() in ['i', 'you', 'he', 'she', 'it', 'we', 'they']:
      tag = 'PRP'
    #possessive pronoun
    elif word.lower() in ['mine', 'his', 'hers', 'yours', 'theirs', 'ours']:
      tag = 'PP$'
    #noun
    elif word[0].isupper():
      tag = "NNP"
    #verb gerund/present participle
    elif word.endswith('ing'):
      tag = "VBG"
    #verb past tense
    elif word.endswith('ed'):
      tag = "VBD"
    #verb past participle
    elif word.endswith('en'):
      tag = "VBN"
    elif word.endswith('es'):
      tag = "VBZ"
    #noun plural
#     elif word.endswith('s'):
#       tag = "NNS"
    elif word.endswith('\'s'):
      tag = "NN$"
    #adverb
    elif word.endswith('ly'):
      tag = "RB"
    #noun
    elif word.endswith('ment') or word.endswith('ness') or word.endswith('ation') or word.endswith('ity'):
      tag = "NN"
    elif word.endswith('ers') or word.endswith('town'):
      tag = "NN"
    #adjective
    elif word.endswith('ful') or word.endswith('able') or word.endswith('ous'):
      tag="JJ"
    #comparative adjective
    elif word.endswith('er'):
      tag = "JJR"
    #superlative adjective
    elif word.endswith('est'):
      tag = "JJS"
    elif word.endswith('.'):
      tag="NN"
#     elif word.islower():
#       tag = "NN"
    else:
      tag = "NN"

    # add the (word, tag) tuple to the list
    tags.append([word, tag])

  return tags

# sentence = "He went to America"
# sentence = "I am going to America at 7 PM"
# sentence = "The quick brown fox jumps over the lazy dog"
# sentence="I am Sunny"
# sentence="Proma is a very good girl"
sentence="lots of people love nahida"
#They love Nahida
tags = pos_tagger(sentence)

length=len(tags)

for i in range(length-2):
    if tags[i][1]=="NN" and tags[i+1][1]=="NN" and tags[i+2][1]=="PREP":
        tags[i+1][1]="VB"
    if tags[i][1]=="PRP" and tags[i+1][1]=="NN" and tags[i+2][1]=="PREP":
        tags[i+1][1]="VB"

for i in range(1, length):
    if tags[i][1]=="NN" and tags[i-1][1]=="NN":
        tags[i-1][1]="JJ"
        
for i in range(1, length):
    if tags[i][1]=="JJ" and tags[i-1][1]=="JJ":
        tags[i-1][1]="RB"
    
print(tags)
```

```python
def pos_tagger(sentence):
  # split the sentence into words
  words = sentence.split()
# initialize list for storing parts of speech tags
  pos_tags = []

  for word in words:
    # check for common nouns (ending in -s or -es)
    if word.endswith('s') or word.endswith('es'):
        pos_tags.append('NN')
    # check for plural proper nouns (ending in -s but not all uppercase)
    elif word.endswith('s') and not word.isupper():
        pos_tags.append('NNP')
    # check for singular proper nouns (first letter uppercase)
    elif word[0].isupper():
        pos_tags.append('NNP')
    # check for determiners (the, a, an)
    elif word.lower() in ['the', 'a', 'an']:
        pos_tags.append('DT')
    # check for personal pronouns (I, you, he, she, it, we, they)
    elif word.lower() in ['i', 'you', 'he', 'she', 'it', 'we', 'they']:
        pos_tags.append('PRP')
    # check for possessive pronouns (my, your, his, her, its, our, their)
    elif word.lower() in ['my', 'your', 'his', 'her', 'its', 'our', 'their']:
        pos_tags.append('PRP$')
    # check for verb forms (ending in -ing or -ed)
    elif word.endswith('ing') or word.endswith('ed'):
        pos_tags.append('VB')
    # check for prepositions (of, in, to, for, with, on, at, from, by)
    elif word.lower() in ['of', 'in', 'to', 'for', 'with', 'on', 'at', 'from', 'by']:
        pos_tags.append('IN')
    # check for conjunctions (and, or, but)
    elif word.lower() in ['and', 'or', 'but']:
        pos_tags.append('CC')
    # check for interjections (oh, wow, huh, etc.)
    elif word.lower() in ['oh', 'wow', 'huh', 'um']:
        pos_tags.append('UH')
    # default to common noun
    else:
        pos_tags.append('NN')
# return list of parts of speech tags
  return pos_tags
sentence = "The quick brown fox jumps over the lazy dog."
tags = pos_tagger(sentence)

print(tags)
```
